### Data Visualization

### Finding the Start up year groups

```python
def start_up_years(row):
    if row['founded_year']>= 1900 and row['founded_year']<=1925:
        row['founded_year_group_list']="<1925"
    elif row['founded_year']>1925 and row['founded_year']<=1950:
        row['founded_year_group_list']="1925-1950"
    elif row['founded_year']>1950 and row['founded_year']<=1975:
        row['founded_year_group_list']="1950-1975"
    elif row['founded_year']>1975 and row['founded_year']<=2000:
        row['founded_year_group_list']="1975-2000"
    elif row['founded_year']>2000:
        row['founded_year_group_list']="2000-2014"
    else:
        row['founded_year_group_list']=""
        
    return row['founded_year_group_list']

df['founded_year_group_list']=df.apply(start_up_years, axis=1)


plt.figure(figsize = (16,7))
sbs.countplot(x = 'founded_year_group_list', data = df.dropna())
plt.title('Distribution of Years in which the Companies were found', fontsize=15)
plt.xlabel('Founded Year Group List', fontsize=14)
plt.ylabel('Count of Number of Companies found in that year range', fontsize=14)
plt.show()

    
```

#### It can be observed that the majority of the companies are established between 2000 and 2014.
#### Very few were established from the years 1902 to 1975. 

### Companies found between the years 2000-2014

```python
df_new = df.loc[df['founded_year'] >= 2000 ]
df_new['founded_year'] = df_new['founded_year'].astype(int)
plt.figure(figsize = (16,7))
sbs.countplot(x = 'founded_year', data = df_new)
plt.title('Companies found in Years from 2000-2014', fontsize=16)
plt.xlabel('Established in Year', fontsize=14)
plt.ylabel('Count of Number of Companies found in that year', fontsize=14)
plt.show()
```

#### The establishment of the finding of the company peaked during the year of 2012, there was an increasing trend from the year 2001 till 2012 then the number of new start ups establishing reduced, indicating there might have been some impact on the market at that time. 

### Top 10 Fields of Startups 

```python
plt.figure(figsize=(16,7))
sbs.countplot(x =' market ', data = df, order=df[' market '].value_counts().iloc[:10].index)
plt.title("Top 10 Fields of Startups", fontsize=16)
plt.xlabel("Field",fontsize=14)
plt.ylabel("Count of Companies based in these Fields",fontsize=14)
plt.xticks(rotation=30)
plt.show()
```

#### Maximum startups are in the field of 'software' technology followed by 'biotechnology' and 'mobile'.

### Distribution of the top 10 countries where the start ups are found in


```python
plt.figure(figsize=(16,7))
g = sbs.countplot(x ='country_code', data = df, order=df['country_code'].value_counts().iloc[:10].index)
plt.title("Countries of Origin", fontsize=16)
plt.xticks(rotation=30)
plt.xlabel("Country Names", fontsize=14)
plt.ylabel("Count of Companies found in these countries", fontsize=14)
plt.show()
```

#### Since the maximum number of companies are bases in USA, it can be only assumed that this dataset was mostly used and updated in the given country. Hence, we go onto see the top state performers of this country. 

### Top States in US

```python
df_USA = df[(df['country_code'] =='USA')]
plt.figure(figsize=(16,7))
g = sbs.countplot(x ='state_code', data = df_USA, order=df['state_code'].value_counts().iloc[:10].index)
plt.title("Top States of US where these Startups are based out of", fontsize=16)
plt.xlabel("Name of the States", fontsize=14)
plt.ylabel("Count of the Companies found in these States", fontsize=14)
plt.xticks(rotation=30)
plt.show()
```

#### The highest number of companies are established in California, so it can be assumed that the presence of 'Silicon Valley' helped play a great role in getting out more start up companies. 


### Company Status

```python
plt.figure(figsize=(16,7))
g= sbs.countplot(x=df['status'], data=df)
plt.title("Status of the Company", fontsize=16)
plt.xlabel("Status", fontsize=14)
plt.ylabel("Count of the Companies", fontsize=14)
plt.show()
```

```python
plt.figure(figsize = (8,8))
df.status.value_counts().plot(kind='pie',shadow=True, explode=(0, 0, 0.15), startangle=90,autopct='%1.1f%%')
plt.title('Status', fontsize=16)
plt.show()
```

#### Pie Chart and Countplot were both used to get the accurate distribution of the status of the companies. As can be noticed, 7.7% of these startups were acquired by the other companies, 86.9% of these startups are still operating or functioning and 5.4% of these startup companies closed down.
